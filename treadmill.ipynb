### **Customer Profiling - Categorization of Users:**

```python
def categorize_age(age):
    if age < 30:
        return "Young"
    elif age < 50:
        return "Middle-aged"
    else:
        return "Elderly"

# Apply the categorization to create a new 'Age_Category' column
af['Age_Category'] = af['Age'].apply(categorize_age)

# View the distribution of customers by age category
age_category_counts = af['Age_Category'].value_counts()
print("Customer Profiling - Categorization of Users:")
print(age_category_counts)
```

We can incearse the sales target to sell the fitness product for Middle - aged and Elderly people

```python
# Define income brackets and labels
income_brackets = [30000, 60000, 90000, 120000]
income_labels = ["Low Income", "Medium Income", "High Income"]

# Categorize customers based on income
af['Income_Category'] = pd.cut(af['Income'], bins=income_brackets, labels=income_labels)

# View the distribution of customers by income category
income_category_counts = af['Income_Category'].value_counts()
print("Customer Profiling - Categorization of Users by Income:")
print(income_category_counts)
```

```python
# Create a contingency table of Income Category vs. Product
income_product_crosstab = pd.crosstab(af['Income_Category'], af['Product'], normalize='index')

# Display the contingency table
print("Contingency Table - Probability of Purchasing Each Product by Income Category:")
print(income_product_crosstab)

# Calculate conditional probabilities
conditional_probabilities = {}
for product in af['Product'].unique():
    for income_category in af['Income_Category'].unique():
        # Check if the values are not NaN
        if pd.notna(product) and pd.notna(income_category):
            probability = income_product_crosstab.loc[income_category, product]
            conditional_probabilities[(income_category, product)] = probability

# Display conditional probabilities
print("\nConditional Probabilities (Income Category -> Product):")
for (income_category, product), probability in conditional_probabilities.items():
    print(f"P(Purchase {product} | {income_category}) = {probability:.2f}")


```
